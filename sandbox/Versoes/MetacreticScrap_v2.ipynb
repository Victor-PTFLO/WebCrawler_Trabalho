```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_name(vreview):
    try:
        r = vreview.find('div', class_='name').find('a').text
    except:
        r = vreview.find('div', class_='review_body').find('span', class_='blurb blurb_expanded')
    return r

def get_rating(vreview):
    try:
        rt = vreview.find('div', class_='review_grade').text.strip( '\n' )
    except:
        rt = None
    return rt

def get_date(vreview):
    return vreview.find('div', class_='date').text

def get_coment(vreview):
    try:
        c = vreview.find('span', class_='blurb blurb_expanded').text
    except: 
        c = vreview.find('div', class_='review_body').find('span', class_='blurb blurb_expanded')
    return c

def get_last_pages(vurl, n_pages = 0, last_page = False):
    if last_page == True:
        bs_last_page = BeautifulSoup(response.text, 'html.parser')
        v_last_page = bs_last_page.find('div', class_='pages').find('li', class_="page last_page").find('a').text
    else:
        v_last_page = n_pages
    return int(v_last_page)
```

```python

url = "https://www.metacritic.com/game/playstation-4/the-last-of-us-remastered/user-reviews"
user_agent = {'User-agent': 'Mozilla/5.0'}

# session = requests.Session()
response = requests.get(url, headers=user_agent)
last_page = get_last_pages(url, last_page=True)
```

```python
startpage = 0
review_lst = []

while startpage <= last_page:
    response = requests.get(url + "?page=" + str(startpage), headers=user_agent)

    #Scrap
    bs = BeautifulSoup(response.text, 'html.parser')
    reviews = bs.find_all('div', class_='review_content')
    review_lst.append(pd.DataFrame([{'name': get_name(row)
                                    ,'date' : get_date(row)
                                    ,'rating': get_rating(row)
                                    ,'coment': get_coment(row)
                                    } for row in reviews]))

    print('-----Loaded Page: ' + str(startpage) + ' -------')
    startpage = startpage+1


df_review = pd.concat(review_lst).dropna()
```
